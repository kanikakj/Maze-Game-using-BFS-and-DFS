```python
import pygame 
from tkinter import *
from PIL import Image,ImageTk
import tkinter as tk
from tkinter import messagebox
import random 
from collections import deque
import math
import time
```

```python
class Spot:
    def __init__(self,i,j):
        self.x=i
        self.y=j
        self.neighbours=[]
        self.prev=None
        self.wall=None
        self.visited=False
        
            
    def show(self,window_sp,color,shape=1):
        if self.wall==True:
            color=(0,0,0)
        if shape==1:
            pygame.draw.rect(window_sp,color,(self.x*w_sp,self.y*h_sp,w_sp-1,h_sp-1))
        elif shape==0:
            pygame.draw.circle(window_sp,color,(self.x*w_sp+w_sp//2,self.y*h_sp+h_sp//2),w_sp//3)
    
    def add_neighbours(self,grid_sp):
        if(self.x > 0):
            self.neighbours.append(grid_sp[self.x-1][self.y])   #left
        if(self.x < cols_sp-1):
            self.neighbours.append(grid_sp[self.x+1][self.y])   #Right
        if(self.y > 0):
            self.neighbours.append(grid_sp[self.x][self.y-1])   #Up
        if(self.y < rows_sp-1):
            self.neighbours.append(grid_sp[self.x][self.y+1])   #Down    
```

```python
def classwall(pos,state):
    
    global window_sp
    global width_sp,height_sp,size_sp,cols_sp,rows_sp,w_sp,h_sp
    global queue_sp,visited_sp,grid_sp,path_sp
    
    x=pos[0]//w_sp
    y=pos[1]//h_sp
    grid_sp[x][y].wall=state
```

```python
def shortest_path():
    time.sleep(0.2)
    r.destroy()
    
    global window_sp
    global width_sp,height_sp,size_sp,cols_sp,rows_sp,w_sp,h_sp
    global queue_sp,visited_sp,grid_sp,path_sp
    
    width_sp=640
    height_sp=480
    size_sp=(width_sp,height_sp)
    
    cols_sp=64
    rows_sp=48
    
    w_sp=width_sp//cols_sp
    h_sp=height_sp//rows_sp
    
    
    queue_sp=deque()    #FIFO
    visited_sp=[]
    grid_sp=[]
    path_sp=[]
    
    
    pygame.init()
    window_sp=pygame.display.set_mode(size_sp)
    pygame.display.set_caption("BFS path searching")
    
    #making grid list
    for i in range(cols_sp):
        arr=[]
        for j in range(rows_sp):
            arr.append(Spot(i,j))
        grid_sp.append(arr)
    
    #adding neighbours
    for i in range(cols_sp):
        for j in range(rows_sp):
            grid_sp[i][j].add_neighbours(grid_sp)
    
    noflag=True
    flag=False
    startflag=False
    
    start=grid_sp[0][0]
    end=grid_sp[cols_sp-cols_sp//2][rows_sp-cols_sp//4]
    start.wall=False
    end.wall=False
    
    queue_sp.append(start)
    start.visited=True
    runing=True
    while runing:
        
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                runing=False
                pygame.quit()
                time.sleep(0.05)
            elif event.type==pygame.MOUSEMOTION:
                if event.buttons[0] or event.buttons[2]:
                    classwall(pygame.mouse.get_pos(),event.buttons[0])
            elif event.type==pygame.MOUSEBUTTONDOWN:
                if event.button in (1,3):
                    classwall(pygame.mouse.get_pos(),event.button==1)
            elif event.type==pygame.KEYDOWN:
                if event.key==pygame.K_RETURN:
                    startflag=True
                
                
        if startflag:
            
            if(len(queue_sp))>0:
                curr=queue_sp.popleft()
                if curr==end:
                    temp=curr
                    while temp.prev:
                        path_sp.append(temp)
                        temp=temp.prev
                    if flag==False:
                        flag=True
                    elif flag:
                        continue
                if flag==False:
                    for i in curr.neighbours:
                        if not i.visited and not i.wall:
                            i.visited=True
                            queue_sp.append(i)
                            i.prev=curr
            else:
                if flag==False and noflag==True:
                    Tk().wm_withdraw()
                    messagebox.showinfo("NO SOLUTION ","There is no Path")
                    runing=False
                    noflag=False
                else:
                    continue
                    
        for i in range(cols_sp):
            for j in range(rows_sp):
                spot=grid_sp[i][j]
                spot.show(window_sp,(39, 174, 96))
                    
                if spot in path_sp:
                    spot.show(window_sp,(46, 204, 113))
                    spot.show(window_sp,(192, 57, 43),0)
                        
                elif spot.visited==True:
                    spot.show(window_sp,(255, 255, 0))
                    
                if spot in queue_sp and flag==False:
                    spot.show(window_sp,(39, 174, 96))
                    spot.show(window_sp,(255, 255, 0),0)
                    
                if spot==start:
                    spot.show(window_sp,(0, 255, 200))

                if spot==end:
                    spot.show(window_sp,(0, 120, 255))
                
        pygame.display.flip() 
    if runing==False:
        main()    
```

```python
def build_grid(x,y):
    global grid_mz
    global WHITE
    for i in range(0,20):
        x=20
        y=y+20
        for j in range(0,20):
            pygame.draw.line(window_mz , WHITE , [x,y],[x+20,y])
            pygame.draw.line(window_mz , WHITE, [x+20,y],[x+20,y+20])
            pygame.draw.line(window_mz, WHITE , [x,y+20],[x+20,y+20])
            pygame.draw.line(window_mz , WHITE , [x,y+20],[x,y])
            grid_mz.append((x,y))
            x=x+20

            
            
def push_up(x,y):
    global w_mz
    global window_mz
    global BLUE
    pygame.draw.rect(window_mz , GREY , (x+1,y-w_mz+1,19,39),0)
    pygame.display.update()
    
def push_down(x,y):
    global w_mz
    global window_mz
    global BLUE
    pygame.draw.rect(window_mz , GREY , (x+1,y+1,19,39),0)
    pygame.display.update()

def push_left(x,y):
    global w_mz
    global BLUE
    global window_mz
    pygame.draw.rect(window_mz , GREY , (x-w_mz+1,y+1,39,19),0)
    pygame.display.update()

def push_right(x,y):
    global w_mz
    global BLUE
    global window_mz
    pygame.draw.rect(window_mz , GREY , (x+1,y+1,39,19),0)
    pygame.display.update()
    
#draw a single width cell
def single_cell(x,y):
    global GREY 
    global window_mz
    pygame.draw.rect(window_mz, BLUE ,(x+1,y+1,18,18),0)
    pygame.display.update()
    

def back_strap(x,y):
    global BLUE
    global window_mz #re-colour path after single_cell has visited
    pygame.draw.rect(window_mz , GREY , (x+1,y+1,18,18),0)
    pygame.display.update()


def solution_cell(x,y):
    global window_mz
    global BLACK
    pygame.draw.rect(window_mz,BLACK,(x+7,y+7,5,5),0)
    pygame.display.update()
    
def plot_route_back(x,y):
    solution_cell(x,y)
    while((x,y)!=(20,20)):
        x,y=solution[x,y]
        solution_cell(x,y)
        time.sleep(0.1)

def crave_maze(x,y):
    single_cell(x,y)
    stack_mz.append((x,y))
    visited_mz.append((x,y))
    
    while(len(stack_mz)>0):
        time.sleep(0.07)
        
        cell=[]
        if (x+w_mz,y) not in visited_mz and (x+w_mz,y) in grid_mz:
            cell.append("right")
            
        if (x-w_mz,y) not in visited_mz and (x-w_mz,y) in grid_mz:
            cell.append("left")
            
        if (x,y+w_mz) not in visited_mz and (x,y+w_mz) in grid_mz:
            cell.append("down")
            
        if (x,y-w_mz) not in visited_mz and (x,y-w_mz) in grid_mz:
            cell.append("up")
            
        if(len(cell)>0):
            cell_chosen=random.choice(cell)
            
            if cell_chosen=="right":
                push_right(x,y)
                solution[(x+w_mz,y)]=x,y
                x=x+w_mz
                visited_mz.append((x,y))
                stack_mz.append((x,y))
                
            if cell_chosen=="left":
                push_left(x,y)
                solution[(x-w_mz,y)]=x,y
                x=x-w_mz
                visited_mz.append((x,y))
                stack_mz.append((x,y))
                
            if cell_chosen=="up":
                push_up(x,y)
                solution[(x,y-w_mz)]=x,y
                y=y-w_mz
                visited_mz.append((x,y))
                stack_mz.append((x,y))
                
            if cell_chosen=="down":
                push_down(x,y)
                solution[(x,y+w_mz)]=x,y
                y=y+w_mz
                visited_mz.append((x,y))
                stack_mz.append((x,y))
                
        else:
            x,y=stack_mz.pop()
            single_cell(x,y)
            time.sleep(0.05)
            back_strap(x,y)
    
```

```python
def maze_game():
    r.destroy()
    
    global window_mz
    global x_mz, y_mz,w_mz
    global grid_mz,visited_mz,stack_mz,solution
    global WHITE,GREY,BLUE,BLACK
    
    # set up pygame window
    WIDTH = 450
    HEIGHT = 450
    FPS = 30

    # Define colours
    WHITE = (255, 255, 255)
    GREY = (192,192,192)
    BLUE = (0, 0, 255)
    BLACK = (0,0,0)
    
    pygame.init()
    window_mz = pygame.display.set_mode((WIDTH,HEIGHT))
    pygame.display.set_caption("DFS Maze Generator ")
    
    x_mz=0
    y_mz=0
    w_mz=20
    grid_mz=[]
    visited_mz=[]
    stack_mz=[]
    solution={}    #dict
    
    x=20
    y=20
    build_grid(20,0)
    
    crave_maze(x,y)
    plot_route_back(400,400)
        
    run=True
    while run:
        
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                run=False
                pygame.quit()
                
                time.sleep(0.01)
    if run==False: 
        main()
```
